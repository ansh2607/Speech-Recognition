```python
import speech_recognition as sr
import webbrowser as wb
import pyttsx3 

# Initialize the recognizer 
r1 = sr.Recognizer()
r2 = sr.Recognizer()

#Function to convert text to speech 
def SpeakText(command): 
    # Initialize the engine 
    engine = pyttsx3.init() 
    engine.say(command) 
    engine.runAndWait() 
```

```python
# Loop infinitely for user to speak 
while(1):
    
    try: 
        # use the microphone as source for input. 
        with sr.Microphone() as source: 

            # wait for a second to let the recognizer adjust the energy threshold based on the surrounding noise level 
            r1.adjust_for_ambient_noise(source, duration=0.2)
            print('speak now..')
            print('--- open google || open youtube || save text ---')

            #listens for the user's input 
            audio = r1.listen(source) 

            # Using google to recognize audio 
            MyText = r1.recognize_google(audio) 
            MyText = MyText.lower() 
            print(MyText) 
            #SpeakText(MyText)
            
            if 'open google' in MyText:
                r2 = sr.Recognizer()
                url = 'https://www.google.com/search?q='
                with sr.Microphone() as source:
                    text = 'Opening google..Speak what you want to search'
                    print(text)
                    SpeakText(text)
                    audio = r2.listen(source)
                    
                    try: 
                        get = r2.recognize_google(audio)
                        print(get)
                        wb.get().open_new(url+get)
                        
                    except sr.RequestError as e: 
                        print("Could not request results; {0}".format(e)) 

                    except sr.UnknownValueError: 
                        print("unknown error occured")
                        
            if 'open youtube' in MyText:
                r2 = sr.Recognizer()
                url = 'https://www.youtube.com/results?search_query='
                with sr.Microphone() as source:
                    text = 'Opening youtube..Speak which video are you interested in?'
                    print(text)
                    SpeakText(text)
                    audio = r2.listen(source)
                    
                    try: 
                        get = r2.recognize_google(audio)
                        print(get)
                        wb.get().open_new(url+get)
                        
                    except sr.RequestError as e: 
                        print("Could not request results; {0}".format(e)) 

                    except sr.UnknownValueError: 
                        print("unknown error occured")
            
            if 'save text' in MyText:
                r2 = sr.Recognizer()
                file1 = open(r"saved_text.txt",'a')
                with sr.Microphone() as source:
                    text = 'Speak the text you want to be saved'
                    print(text)
                    SpeakText(text)
                    audio = r2.listen(source)
                    
                    try: 
                        get = r2.recognize_google(audio)
                        print(get)
                        file1.writelines(get)
                        file1.close()
                        
                    except sr.RequestError as e: 
                        print("Could not request results; {0}".format(e)) 

                    except sr.UnknownValueError: 
                        print("unknown error occured")
                        
            if 'exit' in MyText:
                print('Thank You!!!')
                break

    except sr.RequestError as e: 
        print("Could not request results; {0}".format(e)) 

    except sr.UnknownValueError: 
        print("unknown error occured") 
```
